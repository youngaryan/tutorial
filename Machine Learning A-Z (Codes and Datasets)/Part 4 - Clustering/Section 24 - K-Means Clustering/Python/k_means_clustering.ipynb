# K-Means Clustering

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

## Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
```

## Training the K-Means model on the dataset

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

```python
print(y_kmeans)
```

## Visualising the clusters

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1],color = 'red')
plt.scatter(X[y_kmeans == 1, 0],  X[y_kmeans == 1, 1],color = 'blue')
plt.scatter(X[y_kmeans == 2, 0],  X[y_kmeans == 2, 1],color = 'yellow')
plt.scatter(X[y_kmeans == 3, 0],  X[y_kmeans == 3, 1],color = 'black')
plt.scatter(X[y_kmeans == 4, 0],  X[y_kmeans == 4, 1],color = 'green')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s=400)
```
